```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
baseurl = "https://docsneakers.com/"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'}
```

```python
k = requests.get('https://docsneakers.com/sneakers/').text
soup=BeautifulSoup(k,'html.parser')
productlist = soup.find_all("li",{"class":"product"})
print(productlist)
```

```python
productlinks = []
for product in productlist:
        link = product.find("a",{"class":"button button--small card-figcaption-button"}).get('href')                 
        productlinks.append(baseurl + link)
print(productlinks)
```

```python
productlinks = []
for x in range(1,6):  
  k = requests.get('https://docsneakers.com/sneakers/?sort=featured&page=x'.format(x)).text  
  soup=BeautifulSoup(k,'html.parser')
  productlist = soup.find_all("li",{"class":"product"})

  for product in productlist:
      link = product.find("a",{"class":"button button--small card-figcaption-button"}).get('href')
      productlinks.append(link)
print(productlinks)
```

```python
data=[]
for link in productlinks:
    f = requests.get(link,headers=headers).text
    hun=BeautifulSoup(f,'html.parser')

    try:
      title=hun.find("h1", {"class":"productView-title"}).text.strip().replace("\n","")
    except:
      title=None

    try:
      priceWithTax=hun.find("span", {"class":"price price--withTax"}).text.strip().replace("\n","")
    except:
      priceWithTax=None

    try:
      priceNonSale=hun.find("span", {"class":"price price--non-sale"}).text.strip().strip().replace("\n","")
    except:
      priceNonSale=None
    
    try:
      noOfReviews=hun.find("span", {"class":"productView-reviewLink"}).text.strip().replace("\n","")
    except:
      noOfReviews=None


    product = {"title":title,"priceWithTax":priceWithTax,"priceNonSale":priceNonSale,"reviewLink":noOfReviews}

    data.append(product)

print(data)
```

```python
df = pd.DataFrame(data)

print(df)
```
